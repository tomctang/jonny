# How Dad learn QM


**Learning strategies** - two approaches of learning QM: operator/Algebraic way and calculus/DE way
- Operator way: Start from states ‚à£ùúì‚ü© in Hilbert space and operators ùê¥^ acting on them; Use Dirac notation, commutators, and operator identities instead of solving DEs; algebra-heavy (linear algebra, group theory, operator identities); clean structure (symmetries, conservation laws)
    - Our course is operator-centric, eg. spin systems, four operator identities, Schr√∂dinger factorization and Dirac‚Äôs abstract methodology, angular momentum algebra (rotations, spherical harmonics)
- Calculus way: Start from the Schr√∂dinger equation, heavy use of integrals/DEs/special functions
    - minor calculus elements in our course, eg. wavefunctions in position and momentum space, spherical harmonics, variational method

**Learning outcomes** - upon successfully completing this course, you will:

- M1: Be able to describe how the principle of superposition underlies wave-particle duality and use the quantum superposition of states to analyze properties of spin (Stern-Gerlach experiments) and light (one-, two-, and multiple slits and the Mach-Zehnder interferometer).
- M2: Distinguish between events, alternative ways an event occurs, tagging, and measurement; be able to use the mathematical formalism of quantum mechanics to describe these different phenomena and how they allow for delayed choice experiments.
- M3: Describe how entanglement requires quantum mechanics to violate local realism and how experiments verify that this occurs.
- M4: Explain how an interaction-free experiment allows for quantum seeing in the dark and the details of how such an experiment is carried out.
- M5: Know the four fundamental operator identities of quantum mechanics (Leibniz rule, Hadamard lemma, exponential disentangling identity, and Baker-Campbell-Hausdorff formula), in all of their variants, and be able to use them in quantum calculations.
- M6: Use the Schr√∂dinger factorization method to solve bound-state energy eigenvalue problems (determining the energies algebraically) and couple it with the translation operator (or the subsidiary condition) to find the wavefunctions; explain why the energy levels are discrete and describe how to use the node theorem to verify completeness.
- M7: Use quantum principles employed in the class to explain properties of hydrogen including how to use spectroscopy to measure its mass, the radius of the nucleus, how it is used in radio astronomy, and how the probability distribution of electrons are measured in momentum space.
- M8: Use approximation methods of perturbation theory and the variational method to learn about quantum systems that cannot be solved exactly.
- M9: Calculate how time evolution is employed to describe oscillatory motion of the simple harmonic oscillator and the rotational motion of a nuclear spin; be able to use the Trotter form of the time-ordered product.
- M10: Quantize light, describe what a photon is, how single photons are measured, how to verify you have a single-photon light source, how to squeeze light, and how quantum properties of light are used in gravitational wave detection.

**Notes/Math Definitions**

-   Linear operators preserve scalar multiplication and vector addition, that is:
    - $(\alpha \hat{A} + \beta \hat{B})|\psi\rangle = \alpha \hat{A}|\psi\rangle + \beta \hat{B}|\psi\rangle$
    - $\hat{A}(\alpha|\psi\rangle + \beta|\phi\rangle) = \alpha \hat{A}|\psi\rangle + \beta \hat{A}|\phi\rangle$

-   $\hat{S}_{z}|\uparrow\rangle_{z} = \frac{\hbar}{2}|\uparrow\rangle_{z}$
    $\quad$
    $\hat{S}_{z}|\downarrow\rangle_{z} = -\frac{\hbar}{2}|\downarrow\rangle_{z}$

-   $\hat{S}_{+}|\downarrow\rangle_{z} = \hbar|\uparrow\rangle_{z}$

-   $\hat{S}_{+}|\uparrow\rangle_{z} = 0$

-   $\hat{S}_{-}|\uparrow\rangle_{z} = \hbar|\downarrow\rangle_{z}$ $\quad$
    $\hat{S}_{-}|\downarrow\rangle_{z} = 0$

-   $[\hat{A}, \hat{B}] = \hat{A} \hat{B} - \hat{B} \hat{A}$

-   $[\hat{A}, \hat{B}] = -[\hat{B}, \hat{A}]$

    -   If 2 operators commute, their commutator is equal to 0; 
    - if they do NOT commute, their commutator is nonzero, then we cannot find a complete set of eigenstates that are simultaneous eigenstates of both operators.

-   $\left[\hat{S}_{z}, \hat{S}_{+}\right] = \hbar \hat{S}_{+}$

-   $\left[\hat{S}_{z}, \hat{S}_{-}\right] = -\hbar \hat{S}_{-}$

-   $\left[\hat{S}_{+}, \hat{S}_{-}\right] = 2 \hbar \hat{S}_{z}$

-   $\hat{S}_{x} = \frac{1}{2}\left(\hat{S}_{+} + \hat{S}_{-}\right)$
    $\quad$
    $\hat{S}_{y} = \frac{1}{2 i}\left(\hat{S}_{+} - \hat{S}_{-}\right)$

-   $\hat{S}_{+} = \hat{S}_{x} + i \hat{S}_{y}$ $\quad$
    $\hat{S}_{-} = \hat{S}_{x} - i \hat{S}_{y}$

-   $\left[\hat{S}_{i}, \hat{S}_{j}\right] = i \hbar \sum_{k} \epsilon_{i j k} \hat{S}_{k}$

-   $\hat{S}^{2} = \hat{S}_{x}^{2} + \hat{S}_{y}^{2} + \hat{S}_{z}^{2}$

-   Leibnitz Identity:
    $[\hat{A} \hat{B}, \hat{C}] = \hat{A}[\hat{B}, \hat{C}] + [\hat{A}, \hat{C}] \hat{B}$

-   $\hat{S}_{z} = \begin{pmatrix}
        \frac{\hbar}{2} & 0 \\
        0 & -\frac{\hbar}{2}
        \end{pmatrix}$

-   $\hat{S}_{z}|\uparrow\rangle_{z} = \frac{\hbar}{2}|\uparrow\rangle_{z}$
    $\quad$
    $\hat{S}_{z}|\downarrow\rangle_{z} = -\frac{\hbar}{2}|\downarrow\rangle_{z}$

-   Spin operators in $x, y$ and $z$ are all hermitian, whereas spin
    raising and lowering operators are NOT hermitian.

-   A projection operator $\hat{P}$ satisfies $\hat{P}^{2} = \hat{P}$,
    which is called an idempotent operator. If we have a normalized
    quantum state $|\psi\rangle$, then the operator
    $|\psi\rangle \langle \psi|$ is a projection operator. The sum (or
    integral) over all projection operators (also called 'projectors')
    is the identity matrix. We often use this to 'multiply by 1.'

-   $|\uparrow\rangle_{z} = \frac{1}{\sqrt{2}}\left(|\uparrow\rangle_{x} + |\downarrow\rangle_{x}\right)$

-   The $\otimes$ symbol is called the tensor product; we use this to
    describe quantum states that have different properties, which live
    in different vector spaces. (eg. position and spin, or spin
    and an atomic energy state, and so on\...)

-   $\langle a| \otimes \langle b| \cdot |c\rangle \otimes |d\rangle = \langle a|c\rangle \times \langle b|d\rangle = \langle a|c\rangle \langle b|d\rangle$

-   Superpositioner in Stern Gerlach experiments: transforms
    $|GS\rangle$ to $\frac{1}{\sqrt{2}}(|GS\rangle + |ES\rangle)$

-   De-exciter in Stern Gerlach experiments: behaves as a ground state
    filter

-   Exciter in Stern Gerlach experiments: transforms ground state into
    the excited state

**M4 Notes/Math Definitions**

-   A 50-50 beam splitter in Mach Zehnder Interferometer apparatus
    transforms the incident light $|I\rangle$ into a superposition of
    reflected and transmitted light, denoted by $\frac{1}{\sqrt{2}}(|R\rangle + |T\rangle)$

-   When we make measurements on an arbitrary state $|\psi\rangle$ we
    apply the projection operator (projector) corresponding to our
    measurement onto the state $|\psi\rangle$. Then, to calculate the
    probability, we take the modulus squared of our new state, including
    the projector.

-   When light is polarized in the $\theta$ direction, we express this as:
    $|\theta\rangle = \cos \theta |V\rangle + \sin \theta |H\rangle.$
    Note that $|H\rangle$ and $|V\rangle$ form an orthonormal basis which means that
    $\langle H | H \rangle = \langle V | V \rangle =1$ and $\langle V | H \rangle = \langle H | V \rangle =0$

-   We also define the following state for polarizer problems:
    $|-\theta\rangle = \cos \theta |V\rangle - \sin \theta |H\rangle$

-   If we wish to represent the polarization as a spinor (vector) we
    get: $\begin{pmatrix}
        \cos \theta \\
        \sin \theta
        \end{pmatrix} = \cos \theta |V\rangle + \sin \theta |H\rangle$

-   When a photon goes through a polarizer, it always corresponds to
    being measured, which we accomplish by applying the appropriate
    projection operator onto the photon's state.

-   When a photon goes through a polarization rotator, on the other
    hand, it does not correspond to a measurement. It is just a way to
    manipulate the direction of the polarization of the photon. 
    - We write the action of the rotator as an operator that changes the angle of
    polarization of a polarization state vector as follows:
    $|\theta\rangle \rightarrow \left|\theta + \phi_{\text{rotator}}\right\rangle$
    if the physical rotator rotates the polarization by
    $\phi_{\text{rotator}}$.

-   The binomial coefficient tells us the general formula for $k$
    collections from $n$ objects is:
    $\frac{n!}{k!(n-k)!} = \binom{n}{k}$

-   Binomial theorem:
    $(a + b)^{n} = \sum_{k=0}^{n} \binom{n}{k} a^{k} b^{n-k}$

**M5-6: Notes/Math Definitions** 

Four Fundamental Operator Identities 

-   Leibnitz Rule:
    - $[\hat{A} \hat{B}, \hat{C}] = \hat{A}[\hat{B}, \hat{C}] + [\hat{A}, \hat{C}] \hat{B}$

-   Hadamard:
    - $e^{\hat{A}} \hat{B} e^{-\hat{A}} = \sum_{n=0}^{\infty} \frac{1}{n!} \left[\hat{A}, \left[\hat{A}, \ldots \left[\hat{A}, [\hat{A}, \hat{B}]\right] \ldots \right] \right]_n$

-   Weyl Form of BCH identity:
    - $e^{\hat{A}} e^{\hat{B}} = e^{\hat{A} + \hat{B}} e^{\frac{1}{2} [\hat{A}, \hat{B}]} = e^{\frac{1}{2} [\hat{A}, \hat{B}]} e^{\hat{A} + \hat{B}} = e^{\hat{A} + \hat{B} + \frac{1}{2} [\hat{A}, \hat{B}]}$
    - This holds when $[\hat{A}, \hat{B}]$ commutes with $\hat{A}$ and  $\hat{B}$.

-   General Form of BCH identity:
    - $e^{\hat{A}} e^{\hat{B}} = e^{\left( (\hat{A} + \hat{B}) + \frac{1}{2} [\hat{A}, \hat{B}] + \frac{1}{12} [\hat{A}, [\hat{A}, \hat{B}]] + \frac{1}{12} [\hat{B}, [\hat{B}, \hat{A}]] + \ldots \right)}$
    - This formula comes from the lecture video on BCH.

-   Exponential Disentangling:

    The form of the identity varies depending on your problem and the
    section beginning on Module 6, page 6 provides two examples that are
    worth reviewing. The takeaway is that this procedure lets us rewrite
    an exponential of a matrix in terms of different exponentials of
    matrices that are often easier to calculate and work with.

    Using this identity usually requires remembering the power series expansion for $e^{x}$:
    - $e^{x} = \sum_{n=0}^{\infty} \frac{x^{n}}{n!} = 1 + \frac{x}{1!} + \frac{x^{2}}{2!} + \frac{x^{3}}{3!} + \ldots + \frac{x^{n}}{n!}$
    - You can use a power series to expand exponentials of operators
    (matrices) into a sum with terms that usually either truncate or
    repeat in a useful pattern.

-   Similarity Transformation of Powers (This isn't one of the four
    identities, but it's worth knowing; you proved this on Module 5,
    Page 28.)
    - $\hat{A}^{-1} (\hat{B})^{n} \hat{A} = \left( \hat{A}^{-1} \hat{B} \hat{A} \right)^{n}$

````python
""" Mathematical Unity Between Classical and Quantum Mechanics
1. Poisson brackets ‚Üî Commutators
2. Symplectic geometry ‚Üî Hilbert space geometry
3. Principle of least action ‚Üî Path integrals
4. Symmetries ‚Üî Conservation laws (Noether's theorem)"""
# ============================================================================
# LEVEL 1: POISSON ALGEBRA == OPERATOR ALGEBRA
# ============================================================================
print("LEVEL 1: POISSON BRACKETS ‚Üî COMMUTATORS")
print("="*80)

print("\nClassical mechanics uses Poisson brackets: {f,g} = Œ£·µ¢ (‚àÇf/‚àÇq·µ¢ ‚àÇg/‚àÇp·µ¢ - ‚àÇf/‚àÇp·µ¢ ‚àÇg/‚àÇq·µ¢)")
print("Quantum mechanics uses commutators: [√Ç,BÃÇ] = √ÇBÃÇ - BÃÇ√Ç")
print("Unity: [√Ç,BÃÇ] = i‚Ñè{A,B}ÃÇ  (canonical quantization)")

# Define symbolic variables
var('q p m omega t hbar')

# Classical Hamiltonian for harmonic oscillator
H_classical = p^2/(2*m) + (1/2)*m*omega^2*q^2
print(f"\nClassical Hamiltonian: H = {H_classical}")

# Compute Poisson brackets
def poisson_bracket(f, g, q_vars, p_vars):
    """Compute Poisson bracket {f,g}"""
    result = 0
    for q_i, p_i in zip(q_vars, p_vars):
        result += diff(f, q_i)*diff(g, p_i) - diff(f, p_i)*diff(g, q_i)
    return result

# Fundamental Poisson brackets
pb_qp = poisson_bracket(q, p, [q], [p])
pb_qq = poisson_bracket(q, q, [q], [p])
pb_pp = poisson_bracket(p, p, [q], [p])

print("\nFundamental Poisson brackets:")
print(f"  {{q, p}} = {pb_qp}")
print(f"  {{q, q}} = {pb_qq}")
print(f"  {{p, p}} = {pb_pp}")

# Hamilton's equations from Poisson brackets
dq_dt = poisson_bracket(q, H_classical, [q], [p])
dp_dt = poisson_bracket(p, H_classical, [q], [p])

print("\nHamilton's equations (via Poisson brackets):")
print(f"  dq/dt = {{q, H}} = {dq_dt}")
print(f"  dp/dt = {{p, H}} = {dp_dt}")

# Quantum commutators (symbolic representation)
print("\n" + "-"*80)
print("QUANTUM CORRESPONDENCE:")
print("-"*80)
print("\nCanonical commutation relations:")
print("  [qÃÇ, pÃÇ] = i‚Ñè  ‚ü∫  {q, p} = 1")
print("  [qÃÇ, qÃÇ] = 0   ‚ü∫  {q, q} = 0")
print("  [pÃÇ, pÃÇ] = 0   ‚ü∫  {p, p} = 0")

# Matrix representation of quantum operators (finite dimensional approximation)
print("\nMatrix representation (N=5 harmonic oscillator basis):")
N = 5
# Creation and annihilation operators in matrix form
a_dag = matrix(SR, N, N)
a = matrix(SR, N, N)
for i in range(N-1):
    a_dag[i, i+1] = sqrt(i+1)
    a[i+1, i] = sqrt(i+1)

# Position and momentum operators: qÃÇ = ‚àö(‚Ñè/2mœâ)(a + a‚Ä†), pÃÇ = i‚àö(‚Ñèmœâ/2)(a‚Ä† - a)
q_op = (a_dag + a)
p_op = I*(a_dag - a)

# Compute commutator [qÃÇ, pÃÇ]
commutator_qp = q_op*p_op - p_op*q_op

print(f"\nCommutator [qÃÇ, pÃÇ] (up to constants) has trace: {commutator_qp.trace()}")
print(f"Diagonal elements: {[commutator_qp[i,i] for i in range(3)]}...")

# Uncertainty relation from commutator
print("\n" + "-"*80)
print("UNCERTAINTY PRINCIPLE:")
print("-"*80)
print("From [qÃÇ, pÃÇ] = i‚Ñè, we derive: Œîq¬∑Œîp ‚â• ‚Ñè/2")
print("This is purely quantum and has no classical analog.")
````

# Mathematical Structure Unity: Classical vs Quantum Mechanics

This notebook demonstrates, computationally, the shared mathematical backbone:

- **Classical mechanics:** observables are functions on phase space with a **Poisson bracket**.
- **Quantum mechanics:** observables are operators with a **commutator**.
- **Unity:** both are Lie-algebra dynamics  
  $\dot f = \{f,H\},\qquad \dot{\hat A} = \frac{1}{i\hbar}[\hat A,\hat H].$

## 1) Classical phase space: Poisson bracket + Hamilton equations

````python
var('q p m omega')
R.<q,p> = PolynomialRing(SR, 2)

def poisson(f, g):  # {f,g} = df/dq dg/dp - df/dp dg/dq
    return diff(f,q)*diff(g,p) - diff(f,p)*diff(g,q)

# Hamilton equations: qdot = {q,H}, pdot = {p,H}
H = p^2/(2*m) + (m*omega^2*q^2)/2
qdot = poisson(q, H) ; s(qdot) 
pdot = poisson(p, H) ; s(pdot)

# Jacobi identity check (Poisson bracket is a Lie bracket)
f = q^2*p
g = q*p^2
h = q + p

J = poisson(f, poisson(g,h)) + poisson(g, poisson(h,f)) + poisson(h, poisson(f,g))
s(J)
````

## 2) Quantum: operators + commutators (finite-dimensional truncation)

We model the harmonic oscillator in a truncated Fock space of dimension `N`.
Truncation introduces boundary artifacts near the top state; increasing `N` reduces them.

````python
def annihilation(N):        # a_{n-1,n} = sqrt(n)
    A = matrix(SR, N, N, 0)
    for n in range(1, N):
        A[n-1, n] = sqrt(n)
    return A

def creation(N):
    return annihilation(N).conjugate_transpose()

def comm(A,B):
    return A*B - B*A

# Parameters
N = 10
var('hbar m omega')
a  = annihilation(N)
ad = creation(N)

# x and p in terms of a, a^\dagger
x = sqrt(hbar/(2*m*omega))*(a + ad)
p_op = sqrt(m*omega*hbar/2)*(1/I)*(a - ad)

# Hamiltonian H = p^2/(2m) + (m omega^2 x^2)/2
H = (p_op*p_op)/(2*m) + (m*omega^2*x*x)/2

# Heisenberg equations: xdot = (1/iƒß)[x,H], pdot = (1/iƒß)[p,H]
xdot = (1/(I*hbar))*comm(x, H)
pdot = (1/(I*hbar))*comm(p_op, H)

# Compare with expected: xdot ‚âà p/m, pdot ‚âà -m œâ^2 x (up to truncation effects)
test1 = (xdot - p_op/m)
test2 = (pdot + m*omega^2*x)

subs_dict = {hbar:1, m:1, omega:1}

def frob_norm(M):
    return sqrt(sum((M[i,j].subs(subs_dict))^2 for i in range(N) for j in range(N)))

s(frob_norm(test1))
s(frob_norm(test2))
````

## 3) Canonical commutation: dimensionless quadratures

Define
$ Q=\sqrt{\frac{m\omega}{\hbar}}x,\qquad P=\frac{1}{\sqrt{m\omega\hbar}}p.$
In the infinite-dimensional limit, $[Q,P]=i\mathbf{1}$. In truncation it is close to identity except near the boundary.

````python
Q = sqrt(m*omega/hbar)*x
P = (1/sqrt(m*omega*hbar))*p_op

C = comm(Q, P).subs(subs_dict)
C
````

## 4) Deformation quantization view: Moyal bracket (series)

The Moyal bracket is a deformation of the Poisson bracket:
$\{f,g\}_M = \{f,g\} + O(\hbar^2).$
Below we compute the first correction term via a series expansion.

````python
def moyal_bracket_series(f,g):  # {f,g}_M ‚âà {f,g} - (hbar^2/24) D^3(f,g) + O(hbar^4)
    D1 = poisson(f,g)
    D3 = (diff(f,q,3)*diff(g,p,3)
          - 3*diff(f,q,2,p,1)*diff(g,p,2,q,1)
          + 3*diff(f,q,1,p,2)*diff(g,p,1,q,2)
          - diff(f,p,3)*diff(g,q,3))
    return (D1 - (hbar^2/24)*D3)

f = q^3
g = p^3

s(poisson(f,g))
s(moyal_bracket_series(f,g))
````

## Summary

You have now verified in code:

1. **Classical:** $\{\cdot,\cdot\}$ is a Lie bracket; Hamiltonian dynamics is $\dot f=\{f,H\}$.
2. **Quantum:** $[\cdot,\cdot]$ is a Lie bracket; Heisenberg dynamics is $\dot{\hat A}=\frac{1}{i\hbar}[\hat A,\hat H]$.
3. **Bridge:** dimensionless CCR $[Q,P]\approx i\mathbf{1}$ (exact in the infinite-dimensional limit).
4. **Deformation:** Moyal bracket reproduces Poisson bracket at leading order with $\hbar^2$ corrections.

This is the structural unity: **same dynamical template, different bracket**.
